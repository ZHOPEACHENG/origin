```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['萼片长','萼片宽','花瓣长','花瓣宽','品种'])
```

```python
iris.head()
```

```python
# 1.读取数据并提取花瓣特征列转为NumPy数组；
arr1 = np.array(iris['萼片长'])
arr2 = np.array(iris['萼片宽'])
arr3 = np.array(iris['花瓣长'])
arr4 = np.array(iris['花瓣宽'])
arr5 = np.array(iris['品种'])
print(arr1)
print(arr2)
print(arr3)
print(arr4)
print(arr5)
```

```python
# 2.计算花瓣长均值和方差
print(arr3.mean())
print(arr3.var())
# 花瓣宽的均值和方差
print(arr4.mean())
print(arr4.var())

# 用数组广播计算花瓣长宽比；
a = arr3/arr4
```

```python
# 3.筛选出长宽比大于1.5的样本并统计数量。
a[a>1.5]
```
