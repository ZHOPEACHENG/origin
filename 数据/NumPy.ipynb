```python
arr = np.random.randint(0, 10, size=(3, 3))
print(arr)
print("每列最大值：",np.max(arr, axis=0)) #`axis=0`表示按列求最大值
print("每行最大值：",np.max(arr, axis=1)) #`axis=1`表示按行求最大值
```

```python
np.random.seed(0)
arr = np.random.randint(0, 10, size=(3, 3))
print(arr)
arr[0:2, 0:2] = 0
print(arr)
```
